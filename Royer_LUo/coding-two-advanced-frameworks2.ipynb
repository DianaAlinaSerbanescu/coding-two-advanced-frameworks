Exercise: Scatter Plots

```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
candy_filepath = "./sample_data/candy.csv"

# Fill in the line below to read the file into a variable candy_data
candy_data =pd.read_csv(candy_filepath, index_col="id", parse_dates=True)
```

```python
candy_data.head()
```

```python
popularity_3_musketeers = candy_data.loc[candy_data['competitorname'] == '3 Musketeers', 'winpercent'].iloc[0]
popularity_almond_joy = candy_data.loc[candy_data['competitorname'] == 'Almond Joy', 'winpercent'].iloc[0]

more_popular = '3 Musketeers' if popularity_3_musketeers > popularity_almond_joy else 'Almond Joy'
more_popular
```

```python

sugar_air_heads = candy_data.loc[candy_data['competitorname'] == 'Air Heads', 'sugarpercent'].iloc[0]
sugar_baby_ruth = candy_data.loc[candy_data['competitorname'] == 'Baby Ruth', 'sugarpercent'].iloc[0]

more_sugar = 'Air Heads' if sugar_air_heads > sugar_baby_ruth else 'Baby Ruth'
more_sugar
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(candy_data['sugarpercent'], candy_data['winpercent'])

plt.title('Relationship between Sugar Content and Winning Percentage')
plt.xlabel('Sugar Content (sugarpercent)')
plt.ylabel('Winning Percentage (winpercent)')

plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.regplot(x='sugarpercent', y='winpercent', data=candy_data)

plt.title('Relationship between Sugar Content and Winning Percentage')
plt.xlabel('Sugar Content (sugarpercent)')
plt.ylabel('Winning Percentage (winpercent)')

plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='pricepercent', y='winpercent', hue='chocolate', data=candy_data)

plt.title('Relationship between Price and Winning Percentage with Chocolate as Hue')
plt.xlabel('Price (pricepercent)')
plt.ylabel('Winning Percentage (winpercent)')

plt.show()
```

```python
sns.lmplot(x='pricepercent', y='winpercent', hue='chocolate', data=candy_data)

plt.title('Relationship between Price and Winning Percentage with Chocolate as Hue and Regression Lines')
plt.xlabel('Price (pricepercent)')
plt.ylabel('Winning Percentage (winpercent)')

plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.swarmplot(x='chocolate', y='winpercent', data=candy_data)

plt.title('Relationship between Chocolate and Winning Percentage')
plt.xlabel('Chocolate')
plt.ylabel('Winning Percentage (winpercent)')

plt.show()
```
