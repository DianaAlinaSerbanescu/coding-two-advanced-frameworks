```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
filpath = "/content/sample_data/Marine_CSV_sample.csv"
df = pd.read_csv(filepath)
```

```python
df['Time of Observation'] = pd.to_datetime(df['Time of Observation'], errors='coerce')

wind_speed_data = df[['Time of Observation', 'Wind Speed', 'Air Temperature', 'Air Temperature']]

plt.figure(figsize=(12, 6))

plt.plot(wind_speed_data['Time of Observation'], wind_speed_data['Wind Speed'], label='Average Wind Speed (AWND)', color='blue')

plt.title('Wind Speed Variation Over Time')
plt.xlabel('Time of Observation')
plt.ylabel('Wind Speed (MPH)')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)

plt.show()

```

```python
wind_direction_data = df[['Time of Observation', 'Wind Direction']]

fig, ax1 = plt.subplots(figsize=(12, 6))

color = 'tab:blue'
ax1.set_xlabel('Time of Observation')
ax1.set_ylabel('Wind Speed (MPH)', color=color)
ax1.plot(wind_speed_data['Time of Observation'], wind_speed_data['Wind Speed'], label='Wind Speed', color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()
color = 'tab:red'
ax2.set_ylabel('Wind Direction (Degrees)', color=color)
ax2.bar(wind_direction_data['Time of Observation'], wind_direction_data['Wind Direction'], label='Wind Direction', color=color, alpha=0.5)
ax2.tick_params(axis='y', labelcolor=color)

plt.title('Wind Speed and Direction Variation Over Time')
fig.tight_layout()
plt.show()
```
