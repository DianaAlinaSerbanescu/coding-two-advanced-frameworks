```python
import pandas as pd
reviews = pd.read_csv('/content/sample_data/Pokemon.csv', index_col=0)
import seaborn as sns
```

```python
reviews.head()
```

This dataset was created 7 years ago by ALBERTO BARRADAS. This dataset named Pokemon with stats, Pok√©mon is a Japanese media franchise consisting of video games, animated series and films, a trading card game, and other related media. The franchise takes place in a shared universe in which humans co-exist with creatures known as Pok√©mon, a large variety of species endowed with special powers.

This data set includes 721 Pokemon, including their number, name, first and second type, and basic stats: HP, Attack, Defense, Special Attack, Special Defense, and Speed. It has been of great use when teaching statistics to kids. With certain types you can also give a geeky introduction to machine learning.

This are the raw attributes that are used for calculating how much damage an attack will do in the games.

The data as described is:

ID for each pokemon

Name: Name of each pokemon

Type 1: Each pokemon has a type, this determines weakness/resistance to attacks

Type 2: Some pokemon are dual type and have 2

Total: sum of all stats that come after this, a general guide to how strong a pokemon is

HP: hit points, or health, defines how much damage a pokemon can withstand before fainting

Attack: the base modifier for normal attacks (eg. Scratch, Punch)

Defense: the base damage resistance against normal attacks

SP Atk: special attack, the base modifier for special attacks (e.g. fire blast, bubble beam)

SP Def: the base damage resistance against special attacks
Speed: determines which pokemon attacks first each round

```python
sns.distplot(reviews['HP'])
```

The distribution of the 'HP' columns can be observed and their frequency counts displayed in a histogram. The histogram divides the range of 'HP' values into a number of equal-width intervals and calculates the frequency count within each interval. The height of the bars of the histogram indicates the frequency of the data within the corresponding interval. Also, the distplot() function plots the kernel density estimation curve to show the probability density distribution of the 'HP' values.

```python
sns.kdeplot(reviews.query('Attack').Attack)
```

The distribution of the 'Attack' columns can be observed and their probability densities shown by a kernel density estimation plot.

```python
reviews[reviews['Attack'] < 80]['Attack'].value_counts().sort_index().plot.line()
```

The frequency of values less than 80 in the 'Attack' column can be observed and the distribution is shown in a line graph. The x-axis of the line graph shows the unique values of the 'Attack' column and the y-axis shows the frequency of each unique value. The lines on the line graph connect the frequencies of the different 'Attack' values.

```python
sns.kdeplot(reviews[reviews['HP'] < 80].loc[:, ['Attack', 'Defense']].dropna().sample(200))
```

The bivariate relationship between 'Attack' and 'Defence' can be observed and its distribution can be displayed through a kernel density estimation plot. The kernel density estimation graph displays the probability densities of different 'Attack' and 'Defence' values. The density contours in the graph indicate the joint distribution of these two variables. Sampling 200 samples was done to reduce the amount of data to improve the performance of the plot.

```python
sns.distplot(reviews['Speed'], bins=10, kde=False)
```

```python
sns.distplot(reviews['Speed'], bins=10, kde=False, color='red')
```

With this code, the distribution of the 'Speed' column can be observed and its frequency shown in a histogram. Each column represents a range of intervals, and the height of the column indicates the frequency of data within that interval.

```python
sns.jointplot(x='Defense', y='Speed', data=reviews[reviews['Defense'] < 100])
```

```python
sns.jointplot(x='Defense', y='Speed', data=reviews[reviews['Defense'] < 100], color = 'yellow')
```

```python
sns.jointplot(x='Defense', y='Speed', data=reviews[reviews['Defense'] < 100], kind='hex', gridsize=20, color = 'green')
```

With this code, the relationship between 'Defence' and 'Speed' can be observed and its distribution can be shown in a hexagonal plot. Higher density hexagonal areas indicate more combinations with higher values of 'Defence' and 'Speed', while lower density areas indicate fewer combinations.

```python
df = reviews[reviews['Type 1'].isin(reviews['Type 1'].value_counts().head(5).index)]
sns.boxplot(x='Type 1', y='Attack', data=df)
```

```python
df = reviews[reviews['Type 1'].isin(reviews['Type 1'].value_counts().head(5).index)]
color_palette = sns.color_palette('Set3', n_colors=5)
sns.boxplot(x='Type 1', y='Attack', data=df, palette=color_palette)
```

This code allows you to observe and compare the distribution of Attack values for the five most common Type 1 varieties. Each box has a different colour, making it easier to distinguish between the different varieties.

```python
sns.violinplot(
    x='Type 1',
    y='Speed',
    data=reviews[reviews['Type 1'].isin(reviews['Type 1'].value_counts()[:5].index)]
)
```

```python
color_palette = sns.color_palette('Set3', n_colors=5)
sns.violinplot(
    x='Type 1',
    y='Speed',
    data=reviews[reviews['Type 1'].isin(reviews['Type 1'].value_counts()[:5].index)],
    palette=color_palette
)
```

The 'Speed' distribution of the five most common Type 1 varieties can be observed and visualised in a violin plot. The violin plots show the density of the distribution of 'Speed' values for different varieties, as well as statistical information such as quartiles and medians. Each violin plot has a different colour, making it easier to distinguish between the different varieties.
