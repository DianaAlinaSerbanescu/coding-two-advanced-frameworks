```python
import os
import pandas as pd
pd.plotting.register_matplotlib_converters()

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import numpy as np
import plotly.express as px

# Where to save the figures
PROJECT_ROOT_DIR = "."
CLASS_ID = "Curious explorer"
IMAGES_PATH = os.path.join(PROJECT_ROOT_DIR, "images", CLASS_ID)
os.makedirs(IMAGES_PATH, exist_ok=True)

def save_fig(fig_id, tight_layout=True, fig_extension="png", resolution=300):
    path = os.path.join(IMAGES_PATH, fig_id + "." + fig_extension)
    print("Saving figure", fig_id)
    if tight_layout:
        plt.tight_layout()
    plt.savefig(path, format=fig_extension, dpi=resolution)

print("Setup Complete")
```

```python
CHOSEN_PATH = os.getcwd()+"/sample_data"
DATAFILE = "WineDataset.csv"

def load_chosen_data(chosen_path=CHOSEN_PATH):
    print(CHOSEN_PATH + ".")
    csv_path = os.path.join(chosen_path, DATAFILE)
    return pd.read_csv(csv_path, index_col="Grape", parse_dates=True)

wine_data = load_chosen_data()
```

```python
# Replace 'NV' with NaN in the entire DataFrame
wine_data.replace('NV', np.nan, inplace=True)

# Convert '2020/21' to average year ('2020.5')
wine_data['Vintage'] = wine_data['Vintage'].replace(r'(\d{4})/(\d{2})', r'\1.0', regex=True)

# Extract only the years
wine_data['Vintage'] = wine_data['Vintage'].str.extract(r'(\d{4})').astype(float)

# Handle null values (you can choose the appropriate method)
wine_data.dropna(inplace=True)  # Drop rows with null values

# Check and display column names
print(wine_data.columns)

# Convert 'Vintage' to datetime and set as index
wine_data['Vintage'] = pd.to_datetime(wine_data['Vintage'], format='%Y')
wine_data.set_index('Vintage', inplace=True)

# Time series analysis
# (Add your specific time series analysis code here)

# Example: Number of wines over the years
plt.plot(wine_data.resample('Y').size())
plt.title('Number of Wines Over the Years')
plt.xlabel('Year')
plt.ylabel('Number of Wines')
plt.show()
```

```python
sns.countplot(x='Type', data=wine_data)
plt.title('Distribution of Wine Types')
plt.show()
```

```python
sns.countplot(x='Country', data=wine_data)
plt.title('Distribution of Wines by Country')
plt.xticks(rotation=45)
plt.show()
```

```python
for column in ['Price', 'Capacity', 'ABV']:
    fig = px.histogram(wine_data, x=column, title=f'Histogram of {column}')
    fig.update_traces(marker=dict(color='pink'))
    fig.show()
```

```python
fig = px.box(wine_data, x='Type', y='Price', title='Box Plot: Price by Wine Type')
fig.update_traces(marker=dict(color='green'))
fig.show()
```

```python
fig = px.scatter(wine_data, x='Price', y='ABV', title='Scatter Plot: Price vs. ABV')
fig.update_traces(marker=dict(color='blue'))
fig.show()
```
