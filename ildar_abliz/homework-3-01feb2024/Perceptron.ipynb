```python
import random
import matplotlib.pyplot as plt

def sign(n):
    if n >= 0:
        return 1
    else:
        return -1

class Perceptron:
    def __init__(self):
        self.weights = [random.uniform(-1, 1) for _ in range(2)]
        self.lr = 0.1

    def guess(self, inputs):
        _sum = sum(x * w for x, w in zip(inputs, self.weights))
        output = sign(_sum)
        return output

    def train(self, inputs, target):
        guess = self.guess(inputs)
        error = target - guess

        for i in range(len(self.weights)):
            self.weights[i] += error * inputs[i] * self.lr

class Point:
    def __init__(self):
        self.x = random.uniform(0, 800)
        self.y = random.uniform(0, 800)
        self.label = 1 if self.x > self.y else -1

    def show(self):
        plt.plot(self.x, self.y, 'o', color='white' if self.label == 1 else 'black')

brain = Perceptron()
points = [Point() for _ in range(100)]
trainingIndex = 0

plt.figure(figsize=(8, 8))
plt.plot([0, 800], [0, 800], color='black')

for pt in points:
    pt.show()

for pt in points:
    inputs = [pt.x, pt.y]
    target = pt.label
    guess = brain.guess(inputs)

    if guess == target:
        plt.plot(pt.x, pt.y, 'o', color='green')
    else:
        plt.plot(pt.x, pt.y, 'o', color='red')

training = points[trainingIndex]
inputs = [training.x, training.y]
target = training.label
brain.train(inputs, target)
trainingIndex += 1

if trainingIndex == len(points):
    trainingIndex = 0

plt.show()
```
