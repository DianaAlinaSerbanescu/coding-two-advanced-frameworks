```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline

sns.set_style('darkgrid')# difine the style: whitegrid、ticks、dark、white

data = pd.read_csv('/content/sample_data/fifa.csv',nrows=10) #read the table and limit the row
sns.lineplot(data=data, x='Date', y='ARG', label='ARG')
sns.lineplot(data=data, x='Date', y='BRA', label='BRA')
sns.lineplot(data=data, x='Date', y='ESP', label='ESP')
sns.lineplot(data=data, x='Date', y='FRA', label='FRA')
sns.lineplot(data=data, x='Date', y='GER', label='GER')
sns.lineplot(data=data, x='Date', y='ITA', label='ITA')


plt.legend()
plt.ylabel("fee")
plt.show()
```
