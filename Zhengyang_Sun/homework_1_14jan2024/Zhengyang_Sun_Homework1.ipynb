```python
import os
import pandas as pd
pd.plotting.register_matplotlib_converters()

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

# Where to save the figures
PROJECT_ROOT_DIR = "."
CLASS_ID = "Curious explorer"
IMAGES_PATH = os.path.join(PROJECT_ROOT_DIR, "images", CLASS_ID)
os.makedirs(IMAGES_PATH, exist_ok=True)

def save_fig(fig_id, tight_layout=True, fig_extension="png", resolution=300):
    path = os.path.join(IMAGES_PATH, fig_id + "." + fig_extension)
    print("Saving figure", fig_id)
    if tight_layout:
        plt.tight_layout()
    plt.savefig(path, format=fig_extension, dpi=resolution)

print("Setup Complete")
```

Load and examine the data
Each row in the dataset corresponds to a film. Three film-related data: tomato meter, audience rating, and film running time

```python
# Path of the file to read
mcu_filepath = "/content/sample_data/mcu.csv"

# Read the file into a variable mcu_data
mcu_data = pd.read_csv(mcu_filepath, index_col="tomato_meter")

# Print the first 5 rows of the data
mcu_data.head()
```

Histograms

Suppose we want to create a histogram to see how the audience rating of a film is changing. We can do this using the sns.histplot command.

```python
# Histogram
sns.histplot(mcu_data['audience_score'])
```

```python
# KDE plot
sns.kdeplot(data=mcu_data['audience_score'], shade=True)
```

```python
# 2D KDE plot
sns.jointplot(x=mcu_data['audience_score'], y=mcu_data['movie_duration'], kind="kde")
```

```python
# Histograms for each species
sns.histplot(data=mcu_data, x='audience_score', hue='release_date')

# Add title
plt.title("Histogram of film audience ratings (by release date)")
```

```python
# KDE plots for each species
sns.kdeplot(data=mcu_data, x='audience_score', hue='release_date', shade=True)

# Add title
plt.title("Histogram of film audience ratings (by release date)")
```
